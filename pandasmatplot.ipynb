```python
import pandas as pd
import matplotlib.pyplot as plt
print(pd.__version__)
```

```python
exp1 = pd.read_csv("training.csv")
print(exp1.head())
```

```python
epochs = exp1.epoch

plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.plot(epochs, exp1.loss ,linewidth=4)
plt.plot(epochs, exp1.val_loss,linewidth=4)
plt.title('{} loss'.format("exp1"))
#plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.grid(True)

plt.subplot(132)
plt.plot(epochs, exp1.iou_score,linewidth=4)
plt.plot(epochs, exp1.val_iou_score,linewidth=4)
plt.title('{} IOU score'.format("iou"))
#plt.ylabel('iou_score')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend(['Train', 'Test'], loc='upper left')

plt.subplot(133)
plt.plot(epochs, exp1.f1_score,linewidth=4)
plt.plot(epochs, exp1.val_f1_score,linewidth=4)
plt.title('{} F1 score'.format("f1 score"))
#plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.grid(True)
```
